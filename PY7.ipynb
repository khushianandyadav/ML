```python
import pandas as pd

df=pd.read_csv('hours_vs_score.csv')
```

```python
df
```

```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.scatter(df['Hours_Studied'],df['Exam_Score'])

plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
```

```python
x=df.iloc[:,0]
x
```

```python
y=df.iloc[:,-1]
y
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train.values.reshape(-1, 1),y_train)
```

```python
import numpy as np
lr.predict(np.array(X_test.iloc[0]).reshape(-1,1))
```

```python
plt.plot(df['Hours_Studied'],df['Exam_Score'],'*')
plt.plot(X_train,lr.predict(X_train.values.reshape(-1,1)),color='red')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Exam Score vs Hours Studied with Linear Regression')
plt.grid(True)
plt.show()
```

```python
# y = mx + c

m = lr.coef_
```

```python
c = lr.intercept_
```

```python
y = m * 6.5 + c

y
```
